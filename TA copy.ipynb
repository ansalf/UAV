```python
from collections import defaultdict, deque

def bfs(graph, start, goal):
    # Queue for BFS
    queue = deque([(start, [start])])
    
    # Visited set to keep track of visited nodes
    visited = set()
    
    while queue:
        # Dequeue the current node and its path
        node, path = queue.popleft()
        
        # If the current node is the goal, return the path
        if node == goal:
            return path
        
        # Mark the current node as visited
        visited.add(node)
        
        # Enqueue all unvisited neighbors of the current node
        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))
    
    # If goal is not reachable, return None
    return None

# Contoh penggunaan
# Membangun graf dari data latitude dan longitude dalam DataFrame
def build_graph(df):
    graph = defaultdict(list)
    for i in range(len(df)):
        for j in range(i + 1, len(df)):
            # Misalnya, kita menghitung jarak antara dua titik sebagai bobot pada tepian (edge)
            # Di sini, kita hanya menggunakan jarak euclidean sederhana, namun Anda dapat menggunakan metode lain sesuai kebutuhan
            distance = ((df['latitude'][i] - df['latitude'][j]) ** 2 + (df['longitude'][i] - df['longitude'][j]) ** 2) ** 0.5
            # Menambahkan tepian (edge) dua arah antara dua titik
            graph[i].append((j, distance))
            graph[j].append((i, distance))
    return graph

# Membaca file CSV
df = pd.read_csv('dataset/fire_archive_M6_96619_10data.csv')

# Membangun graf dari DataFrame
graph = build_graph(df)

# Menentukan simpul start dan goal
start = 0
goal = 9

# Menjalankan algoritma BFS untuk mencari jalur yang optimal dari simpul start ke simpul goal
path = bfs(graph, start, goal)

if path:
    print("Jalur optimal dari simpul {} ke simpul {}:".format(start, goal), path)
else:
    print("Tidak ada jalur yang ditemukan dari simpul {} ke simpul {}.".format(start, goal))
```

```python
import pandas as pd
from collections import defaultdict, deque

def bfs(graph, start, goal):
    # Queue for BFS
    queue = deque([(start, [start])])
    
    # Visited set to keep track of visited nodes
    visited = set()
    
    while queue:
        # Dequeue the current node and its path
        node, path = queue.popleft()
        
        # If the current node is the goal, return the path
        if node == goal:
            return path
        
        # Mark the current node as visited
        visited.add(node)
        
        # Enqueue all unvisited neighbors of the current node
        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))
    
    # If goal is not reachable, return None
    return None

# Function to build a graph from latitude and longitude data in DataFrame
def build_graph(df):
    graph = defaultdict(list)
    for i in range(len(df)):
        for j in range(i + 1, len(df)):
            # Calculate distance between two points as edge weight
            distance = ((df['latitude'][i] - df['latitude'][j]) ** 2 + (df['longitude'][i] - df['longitude'][j]) ** 2) ** 0.5
            # Add two-way edge between two points
            graph[i].append(j)
            graph[j].append(i)
    return graph

# Read CSV file
df = pd.read_csv('dataset/fire_archive_M6_96619_10data.csv')

# Build graph from DataFrame
graph = build_graph(df)

# Define start and goal nodes
start = 0
goal = 9

# Run BFS algorithm to find optimal path from start node to goal node
path = bfs(graph, start, goal)

if path:
    print("Optimal path from node {} to node {}:".format(start, goal), path)
else:
    print("No path found from node {} to node {}.".format(start, goal))
```

```python
import pandas as pd
from collections import defaultdict, deque

def bfs(graph, start, end):
    queue = deque([start])
    visited = {start}
    distance = {start: 0}
    path = {start: [start]}
    
    while queue:
        node = queue.popleft()
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                distance[neighbor] = distance[node] + 1
                path[neighbor] = path[node] + [neighbor]
                queue.append(neighbor)
    
    if end not in distance:
        print(f"Node {end} tidak dapat dijangkau dari {start}")
        return None, None
    
    return distance[end], path[end]

# Sample DataFrame (replace this with your actual DataFrame)
data = {
    "latitude": [-11.8070, -11.7924, -12.8398, -14.4306, -12.4953, -12.6191, -14.3655, -14.3195, -13.1654, -11.5473],
    "longitude": [142.0583, 142.0850, 132.8744, 143.3035, 131.4897, 142.1998, 143.5682, 143.5198, 141.9715, 132.6796],
    "brightness": [313.0, 319.3, 311.6, 310.1, 310.3, 314.8, 305.4, 322.9, 317.2, 311.5]
}
df = pd.DataFrame(data)

# Build graph based on DataFrame
def build_graph(df):
    graph = defaultdict(list)
    for i, row in df.iterrows():
        for j, other_row in df.iterrows():
            # We can define the conditions under which two nodes are connected here
            if i != j and abs(row['latitude'] - other_row['latitude']) < 1 and abs(row['longitude'] - other_row['longitude']) < 1:
                graph[i].append(j)
    return graph

graph = build_graph(df)

# Define start and end nodes
start = 0  # For example
end = 9    # For example

# Run BFS algorithm
distance, path = bfs(graph, start, end)

# Print results
if distance is not None:
    print(f"Jarak dari node {start} ke node {end}: {distance}")
    print(f"Rute yang dilewati: {path}")
```
