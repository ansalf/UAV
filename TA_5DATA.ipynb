```python
import csv
import math

class Node:
    def __init__(self, id, latitude, longitude):
        self.id = id
        self.latitude = latitude
        self.longitude = longitude
        self.neighbors = {}
        self.g_cost = math.inf
        self.h_cost = None
        self.parent = None
    
    def set_h_cost(self, target):
        # Menggunakan jarak Euclidean sebagai heuristik
        self.h_cost = math.sqrt((self.latitude - target.latitude)**2 + (self.longitude - target.longitude)**2)

class AStar:
    def __init__(self, nodes):
        self.nodes = nodes

    def find_path(self, start, target):
        open_set = set()
        closed_set = set()
        current = start
        open_set.add(current)

        while open_set:
            current = min(open_set, key=lambda x: x.g_cost + x.h_cost)
            if current == target:
                path = []
                while current.parent:
                    path.append(current.id)
                    current = current.parent
                path.append(start.id)
                path.reverse()
                return path

            open_set.remove(current)
            closed_set.add(current)

            for neighbor_id, cost in current.neighbors.items():
                neighbor = self.nodes[neighbor_id]
                if neighbor in closed_set:
                    continue
                
                tentative_g_cost = current.g_cost + cost
                if tentative_g_cost < neighbor.g_cost:
                    neighbor.parent = current
                    neighbor.g_cost = tentative_g_cost
                    if neighbor not in open_set:
                        open_set.add(neighbor)
                        neighbor.set_h_cost(target)
        return None

def read_csv(filename):
    nodes = {}
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            id = len(nodes)
            node = Node(id, float(row['latitude']), float(row['longitude']))
            nodes[id] = node
    return nodes

def add_edges(nodes):
    for id1, node1 in nodes.items():
        for id2, node2 in nodes.items():
            if id1 != id2:
                distance = math.sqrt((node1.latitude - node2.latitude)**2 + (node1.longitude - node2.longitude)**2)
                node1.neighbors[id2] = distance
    return nodes

def main():
    filename = 'dataset/fire_archive_M6_96619_5data.csv'
    nodes = read_csv(filename)
    nodes = add_edges(nodes)
    start_id = 0
    target_id = 4
    start_node = nodes[start_id]
    target_node = nodes[target_id]

    astar = AStar(nodes)
    path = astar.find_path(start_node, target_node)

    if path:
        print("Path found:", path)
    else:
        print("Path not found")

if __name__ == "__main__":
    main()
```
